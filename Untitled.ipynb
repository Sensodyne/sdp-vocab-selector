```python
oplist = op['vocabulary'].tolist()
aclist = ac['vocabulary'].tolist()
calist = ca['vocabulary'].tolist()
delist = de['vocabulary'].tolist()
```

```python
overlapped_op_ac = (set(oplist) & set(aclist))
overlapped_op_ca = (set(oplist) & set(calist))
overlapped_op_de = (set(oplist) & set(delist))
overlapped_ca_de = (set(calist) & set(delist))
overlapped_ca_ac = (set(calist) & set(aclist))
overlapped_de_ac = (set(delist) & set(aclist))


# overlapped_whole = (set(active_most_frequent_vocab) & set(operation_most_frequent_vocab) &
#                   set(cancelled_most_frequent_vocab) & set(delayed_most_frequent_vocab))
overlapped_op_ac
```

```python
def exclude_overlapped_ones_simple(target_vocab:list, target:pd.core.series.Series, status1:set, status2:set, status3:set):
    excluded_indexer = list(set(target_vocab) - status1 - status2 - status3)
    excluded = target.loc[excluded_indexer]
    return excluded
```

```python
op_result = exclude_overlapped_ones_simple(oplist, op_series, overlapped_op_ac, overlapped_op_ca, overlapped_op_de)
```
